
**Tokenization** in Natural Language Processing (NLP) is the process of breaking down a text or a sequence of characters into smaller units, typically words or subwords, called tokens. Tokens are the fundamental building blocks for many NLP tasks and enable the computer to understand and process human language.

```python
text = "On 24 September 1493, Columbus sailed from Cádiz with 17 ships, and supplies to establish permanent colonies in the Americas"
tokens = text.split()  # Split by whitespace
print(tokens)
```

```python
import nltk
nltk.download('punkt')  # Download the Punkt tokenizer models (if not already downloaded)
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)
print(tokens)
```

```python
import spacy

nlp = spacy.load('en_core_web_sm')  # Load the English language model
doc = nlp(text)

tokens = [token.text for token in doc]
print(tokens)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
tokens = tokenizer.tokenize(text)
print(tokens)
```

**Stop words** are common words in a language that are often considered to be of little value in text analysis and natural language processing (NLP) tasks because they occur frequently in text and don't carry significant meaning on their own. Examples of stop words in English include "the," "and," "is," "in," "to," "of," "a," and many others.

```python
import nltk
nltk.download('stopwords')  # Download the stopwords corpus (if not already downloaded)
from nltk.corpus import stopwords

# Get the list of English stopwords
stop_words = set(stopwords.words('english'))

print(stop_words)
```

```python
import spacy

nlp = spacy.load('en_core_web_sm')  # Load the English language model
stop_words = spacy.lang.en.stop_words.STOP_WORDS

print(stop_words)
```

**Stemming** is a text preprocessing technique in Natural Language Processing (NLP) that involves reducing words to their root or base form. The purpose of stemming is to normalize words with the same meaning but different inflections or derivations to their common base form. Stemming helps in reducing the dimensionality of the vocabulary and improving the efficiency of text analysis and retrieval.

```python
import nltk
nltk.download('punkt')  # Download the Punkt tokenizer models (if not already downloaded)
from nltk.stem import PorterStemmer

# Create a stemmer object
stemmer = PorterStemmer()

# Sample words for stemming
words = ["jumping", "jumps", "jumped", "jumper", "jump"]

# Stem the words
stemmed_words = [stemmer.stem(word) for word in words]

print(stemmed_words)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")  # Load the English language model
text = "jumping jumps jumped jumper jump"

doc = nlp(text)

stemmed_words = [token.lemma_ for token in doc]
print(stemmed_words)
```


**Lemmatization** is a text preprocessing technique in Natural Language Processing (NLP) that involves reducing words to their base or dictionary form, known as a lemma. Unlike stemming, which simply removes prefixes or suffixes from words to get to the root form, lemmatization takes into account the context and grammar of the words to produce valid and meaningful lemmas.

```python
import nltk
from nltk.stem.porter import PorterStemmer
porter_stemmer  = PorterStemmer()
text = "studies studying cries cry"
tokenization = nltk.word_tokenize(text)
for w in tokenization:
  print("Stemming for {} is {}".format(w,porter_stemmer.stem(w)))
```

**The Bag of Words (BoW)** is a common text representation technique in Natural Language Processing (NLP). It represents a text document as a "bag" or a collection of its individual words, disregarding the order and structure of the words but keeping track of their frequency or occurrence. The main idea behind BoW is to convert text data into numerical format, which can be used as input for machine learning algorithms.

```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample text documents
documents = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?"
]

# Create an instance of CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the documents into a BoW matrix
X = vectorizer.fit_transform(documents)

# Get the feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Convert the BoW matrix to an array and print it
bow_array = X.toarray()
print(bow_array)

# Print the feature names
print(feature_names)
```

**Term Frequency (TF):**
Term Frequency is a measure of how frequently a term (word) appears in a document. It is calculated as the count of a term's occurrence in a document divided by the total number of terms in that document.

```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample text documents
documents = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?"
]

# Create an instance of CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the documents into a BoW matrix
X = vectorizer.fit_transform(documents)

# Calculate Term Frequency (TF)
tf = X.toarray()
print("Term Frequency (TF):\n", tf)
```

**Inverse Document Frequency (IDF):**
Inverse Document Frequency is a measure of how important or unique a term is across a collection of documents. It is calculated as the logarithm of the total number of documents divided by the number of documents containing the term.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create an instance of TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the documents into a TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Calculate Inverse Document Frequency (IDF)
idf = tfidf_vectorizer.idf_
print("Inverse Document Frequency (IDF):\n", idf)
```

**Term Frequency-Inverse Document Frequency (TF-IDF):**

TF-IDF is a combination of TF and IDF. It represents the importance of a term in a document relative to its importance in the entire corpus. TF-IDF is calculated by multiplying the TF of a term in a document by the IDF of that term.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create an instance of TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the documents into a TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Get the TF-IDF values
tfidf_values = tfidf_matrix.toarray()
print("TF-IDF Values:\n", tfidf_values)
```

Tokenization Examples

## 1-gram Tokenization (Individual Words):

```python
import nltk
from nltk.util import ngrams

# Sample text
text = "On 24 September 1493, Columbus sailed from Cádiz with 17 ships, and supplies to establish permanent colonies in the Americas"

# Tokenize into 1-grams (individual words)
tokens = nltk.word_tokenize(text)
print(tokens)
```

## 2-gram Tokenization (Bigrams):

```python
# Tokenize into 2-grams (bigrams)
tokens = nltk.word_tokenize(text)
bigrams = list(ngrams(tokens, 2))
print(bigrams)
```

## 4-gram Tokenization (Bigrams):

```python
# Tokenize into 4-grams
tokens = nltk.word_tokenize(text)
n_grams = list(ngrams(tokens, 4))
print(n_grams)
```
